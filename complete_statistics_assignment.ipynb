## Questions & Answers

### Q1. What are the three measures of central tendency?
**Answer:** Mean, Median, and Mode.

### Q2. Difference between mean, median, and mode?
**Answer:** Mean = average, Median = middle value, Mode = most frequent value.

### Q3. Measures for given data

### Q5. Measures of dispersion
**Answer:** Range, variance, and standard deviation describe spread of data.

### Q6. What is a Venn diagram?
**Answer:** A diagram showing relationships between sets using circles.

### Q7. A = (2,3,4,5,6,7), B = (0,2,6,8,10)

### Q8. Skewness
**Answer:** Skewness shows asymmetry of data distribution.

### Q9. Right skewed data
**Answer:** Mean > Median.

### Q10. Covariance vs Correlation
**Answer:** Covariance shows direction, correlation shows strength and direction.

### Q11. Sample mean formula
**Answer:** x̄ = Σx / n

### Q12. Normal distribution
**Answer:** Mean = Median = Mode.

### Q13. Covariance vs correlation (again)
**Answer:** Correlation is standardized form of covariance.

### Q14. Outliers effect
**Answer:** Outliers affect mean and standard deviation significantly.
